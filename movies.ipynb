```python
#importing the necessary libraries
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns
```

```python
# creating a copy before beginning data cleaning
movies1 = movies.copy(deep=True)
movies1.head(2)
```

```python
# checking unique variable
for i in movies1:
  colval = movies1[i].unique()
  print(f"Column name: {i} \n : {colval}")
```

```python
# checking columns again
movies1.columns
```

```python
# convert the columns into lowercase
movies1.columns = movies1.columns.str.lower()
# confirming the changes
movies1.columns
```

```python
# Clean up a value in director facebook likes
movies1['director_facebook_likes'] = movies1['director_facebook_likes'].str.replace('"', '', regex=False)
```

```python
# Rename column titles
movies1.rename(columns={"title_year.1":"title_year_1"}, inplace=True)
movies1.rename(columns={"cast_total_facebook_likes" : "total_cast_facebook_likes"}, inplace=True)
movies1.columns
```

```python
# checking missing values
movies1.isnull().sum()
```

```python
movies1
```

```python
# filling in missing values
movies1.fillna({'duration': movies1['duration']. median()},inplace=True)
movies1.fillna({'director_facebook_likes' : movies1['director_facebook_likes'].median()},inplace=True)
movies1.fillna({'facenumber_in_poster' : movies1['facenumber_in_poster'].median()}, inplace=True)
movies1.fillna({'total_cast_facebook_likes' : movies1['total_cast_facebook_likes'].median()}, inplace=True)
movies1.fillna({'num_voted_users': movies1['num_voted_users'].median()}, inplace=True)
movies1.fillna({'actor_2_facebook_likes': movies1['actor_2_facebook_likes'].median()}, inplace=True)
movies1.fillna({'title_year_1' : movies1['title_year']}, inplace=True)
```

```python
movies1.isnull().sum()
```

```python
# converting data into their right datatypes
movies1 = movies1.astype({'actor_2_facebook_likes':int, 'title_year_1': int, 'total_cast_facebook_likes':int, 'num_voted_users':int, 'director_facebook_likes':int})
movies1.info()
```

```python
# dealing with the NaN value
movies1['director_facebook_likes'] = pd.to_numeric(movies1['director_facebook_likes'], errors='coerce').fillna(0).astype(int)
```

```python
# checking for duplicates
movies1.duplicated().sum()
```

```python
# drop duplicates
movies1.drop_duplicates(inplace=True)
# check again for duplicates
movies1.duplicated().sum()
```

```python
# cleaning the movie title column
movies1['movie_title'] = movies1['movie_title'].str[:-2]
movies1.head()
```

```python
movies1
```

```python
# outlier handling
movies1['log_budget'] = np.log1p(movies1['budget'])
movies1['log_gross'] = np.log1p(movies1['gross'])
movies1['log_num_voted_users'] = np.log1p(movies1['num_voted_users'])
```
